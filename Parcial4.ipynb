```python
##Punto 1. a) implemente el método de Euler hacia adelante (Forward) y el método de RungeKutta de 4to orden para hallar x1(t), 
#x2(t), x3(t), de 0 a 2880 min con pasos de h=1 min.

##Método de Euler hacia adelante
import numpy as np 
import matplotlib.pyplot as plt

def dx1(x3,x1):
    return (0.001*x3)-(0.001*x1)+0.125
def dx2(x1,x2):
    return (0.001*x1)-(0.001*x2)
def dx3(x2,x3):
    return (0.001*x2)-(0.001*x3)

t0 = 0
tf = 2880
h = 1
dx10 = 0.0
dx20 = 0.0
dx30 = 0.0
T = np.arange(t0,tf+1,h)
x1euler = np.zeros(len(T))
x1euler[0] = dx10
x2euler = np.zeros(len(T))
x2euler[0] = dx20
x3euler = np.zeros(len(T))
x3euler[0] = dx30
x1 = 0
x2 = 0
x3 = 0
t = t0
z = 1

while 1:
    if z >= len(T):
        break
    dX1 = dx10 + h*dx1(x3,x1)
    x1euler[z] = dX1
    dX2 = dx20 + h*dx2(x1,x2)
    x2euler[z] = dX2
    dX3 = dx30 + h*dx3(x2,x3)
    x3euler[z] = dX3
    t = t+h
    dx10 = dX1
    dx20 = dX2
    dx30 = dX3
    x1 = dX1
    x2 = dX2
    x3 = dX3
    z +=1


print(x1euler[2880])
print(x2euler[2880])
print(x3euler[2880])

plt.figure()
plt.plot(T,x1euler,'b')
plt.plot(T,x2euler,'r')
plt.plot(T,x3euler,'g')
```

```python
##Método de RungeKutta de 4to orden
import numpy as np 
import matplotlib.pyplot as plt

def dx1(x3,x1):
    return (0.001*x3)-(0.001*x1)+0.125
def dx2(x1,x2):
    return (0.001*x1)-(0.001*x2)
def dx3(x2,x3):
    return (0.001*x2)-(0.001*x3)

t0 = 0
tf = 2880
h = 1
dx10 = 0.0
dx20 = 0.0
dx30 = 0.0
T = np.arange(t0,tf+1,h)
x1RK = np.zeros(len(T))
x1RK[0] = dx10
x2RK = np.zeros(len(T))
x2RK[0] = dx20
x3RK = np.zeros(len(T))
x3RK[0] = dx30
x1 = 0
x2 = 0
x3 = 0
t = t0
z = 1

while 1:
    if z >= len(T):
        break
    k1x1 = dx1(x3,x1)
    k2x1 = dx1(x3+0.5*h*k1x1,x1+0.5*h)
    k3x1 = dx1(x3+0.5*h*k2x1,x1+0.5*h)
    k4x1 = dx1(x3+h*k3x1,x1)
    dX1 = x1+((h/6)*(k1x1+2*k2x1+2*k3x1+k4x1))
    x1RK[z] = dX1
    dx10 = dX1
    x1 = dX1
    
    k1x2 = dx2(x1,x2)
    k2x2 = dx2(x1+0.5*h*k1x2,x2+0.5*h)
    k3x2 = dx2(x1+0.5*h*k2x2,x2+0.5*h)
    k4x2 = dx2(x1+h*k3x2,x2)
    dX2 = x2+((h/6)*(k1x2+2*k2x2+2*k3x2+k4x2))
    x2RK[z] = dX2
    dx20 = dX2
    x2 = dX2
    
    k1x3 = dx3(x2,x3)
    k2x3 = dx3(x2+0.5*h*k1x3,x3+0.5*h)
    k3x3 = dx3(x2+0.5*h*k2x3,x3+0.5*h)
    k4x3 = dx3(x2+h*k3x3,x3)
    dX3 = x3+((h/6)*(k1x3+2*k2x3+2*k3x3+k4x3))
    x3RK[z] = dX3
    dx30 = dX3
    x3 = dX3
    
    t = t+h
    z +=1
    
print(x1RK[2880])
print(x2RK[2880])
print(x3RK[2880])

plt.figure()
plt.plot(T,x1RK,'b')
plt.plot(T,x2RK,'r')
plt.plot(T,x3RK,'g')
```
