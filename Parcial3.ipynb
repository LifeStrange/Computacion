```python
##Solucion de ecuaciones no lineales. Metodos Iterativos 
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return (-1*((3*x)**(1/5)))+((x**3)*np.cos(3*x))+(4*(x**2))-7
def df1(x):
    return (-(1/5)*3*((3*x)**(-4/5)))+(3*(x**2)*np.cos(3*x))+((x**3)*-3*np.sin(3*x))+(8*x)
def g1(x):
    return np.sqrt((1/4)*(7+((3*x)**(1/5))-((x**3)*np.cos(3*x))))

x = np.arange(0,2,0.001)

plt.figure()
plt.plot(x,f1(x))
plt.figure()
plt.plot(x,x)
plt.plot(x,g1(x))

Tol = 10.0**(-10)
x0 = 2.0
x1 = 2.0

def met(x0,x1,Tol,f1,df1):
    x2pre = 0.0
    iter = 0
    x2array = []
    while 1:
#         x2 = (x0+x1)/2        ##Bisección
#         x2 = x0-(f1(x0)*(x1-x0)/(f1(x1)-f1(x0))) ## Falsa posicion
#         x2 = x0-(f1(x0)/df1(x0))     ## Newton
#         x2 = ((x1-x0)/(f1(x1)-f1(x0)))*f1(x0)
        x2 = g1(x0)
        if np.abs(x2-x2pre) <= Tol:
            break
        if np.abs(f1(x2)) <= Tol:
            break
#         if f1(x0)<0 and f1(x2)<0 or f1(x0)>0 and f1(x2)>0: ## Descomentar para metodo de Bisección
#             x0=x2
#         else:                                              ## Descomentar para metodo de Bisección
#             x1=x2
        else:
            x1 = x0
            x0 = x2
        x2pre = x2
        iter += 1
        x2array = np.append(x2array,x2)
    return x2array
X = met(x0,x1,Tol,f1,df1)

print(X)
print(f1(X))
        
```
